```python
import numpy as np
from fractions import Fraction as F 
class Adeline(object):
    def __init__(self,np_size,size):
#         self.berat = np.zeros((size,np_size))
        self.berat = np.random.rand(size,np_size)
        self.bias = np.zeros(size)

    def get_berat(self):
        return (self.berat,self.bias)    
    
    def fungsi_aktivasi(self, v, threshold):
        if v > threshold:
            return 1
        else:
            return -1
    
    def train(self,train_data,train_target,alpha,threshold):
        max_error = threshold
        epoch = 1
        while max_error >= threshold:
            print("EPOCH ", epoch)
            for i in range(len(train_data)):
                v = np.zeros(len(train_target))
                y = np.zeros(len(train_target))
                for j in range(len(train_target)):
#                     v[j] = F(np.dot(train_data[i], self.berat[j]) + self.bias[j])
                    v[j] = (np.dot(train_data[i], self.berat[j]) + self.bias[j])

                    y[j] = self.fungsi_aktivasi(v[j], threshold)
                    print("v",j,": ",v[j])
                    print("y",j,": ",y[j])
                    temp = self.berat[j]
                    print("temp",temp)
                    self.berat[j] = self.berat[j] + (alpha * (train_target[i][j] - v[j])*train_data[i])
                    self.bias[j] = self.bias[j] + (alpha * (train_target[i][j] - v[j]))
                    max_error = np.max(np.append(v[j],max_error))
#                     max_error = np.sum(y[j]-v[j]) * np.sum(y[j]-v[j]) 
            epoch += 1
            (PrintBerat,PrintBias) = self.get_berat()
            print("error",max_error)
            print("berat: ",PrintBerat)
            print("bias: ",PrintBias)
            print('------------------------------------------------------------------------------------')

    def test(self, test_data, threshold):
        v = np.zeros(len(self.berat))
        y = np.zeros(len(self.berat))
        for i in range(len(self.berat)):
            v[i] = np.dot(test_data, self.berat[i]) + self.bias[i]
            y[i] = self.fungsi_aktivasi(v[i], threshold)
        return y
    
```
