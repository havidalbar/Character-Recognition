# Membuat Logic Klasifikasi Single Layer Perceptron

```python
import numpy as np

class SingleLayerPerceptron(object):
    def __init__(self,np_size,size):
        self.berat = np.zeros((size,np_size))
        self.bias = np.zeros(size)

    def get_berat(self):
        return (self.berat,self.bias)    
    
    def fungsi_aktivasi(self, v, threshold):
        if threshold >= v and v >= -(threshold):
            return 0
        elif v > threshold:
            return 1
        else:
            return -1
    
    def train(self,train_data,train_target,threshold):
        epoch = 1
        stop = False
        while stop is False:
            stop = True
            print("EPOCH ", epoch,end = '\n' )
            for i in range(len(train_data)):
                v = np.zeros(len(train_target))
                y = np.zeros(len(train_target))
                for j in range(len(train_target)):
                    v[j] = np.dot(train_data[i], self.berat[j]) + self.bias[j]
                    y[j] = self.fungsi_aktivasi(v[j], threshold)
                    print("v index ",j,": ",v[j])
                    print("y index ",j,": ",y[j])
                    if y[j] != train_target[i][j]:
                        self.berat[j] = self.berat[j] + (train_target[i][j] * train_data[i])
                        self.bias[j] = self.bias[j] + train_target[i][j]
                        stop = False
                    print('------------------------------------------------------------------------------------')

            epoch += 1
            (PrintBerat,PrintBias) = self.get_berat()
            print("berat: ",PrintBerat)
            print("bias: ",PrintBias)
            print("kondisi berhenti: ", stop)
            print('====================================================================================', end = '\n\n')

    def test(self, test_data, threshold):
        v = np.zeros(len(self.berat))
        y = np.zeros(len(self.berat))
        for i in range(len(self.berat)):
            v[i] = np.dot(test_data, self.berat[i]) + self.bias[i]
            y[i] = self.fungsi_aktivasi(v[i], threshold)
        return y
```

# Membuat Method Untuk Mengkonversi Inputan Gambar Menjadi Numeric

```python
def huruf2array(huruf):
    huruf=huruf.replace("#","1,").replace(".","0,").replace("\n",'')
    arr = np.fromstring(huruf[:-1],dtype=int,sep=',')
    return arr

file = open("data/input_latih1_huruf_a.txt", 'r')
x = file.read()
file.close()
print("Hasil Numeric Data Latih Huruf A: ")
print(huruf2array(x))
```
